```python
import pandas as pd
import json
from bs4 import BeautifulSoup
import pandas as pd
import requests

df = pd.read_csv("./unesco_sites.csv")
df["short_description_en"] = (df["short_description_en"]
    .astype(str)
    .str.replace("<p>", "", regex=False)
    .str.replace("</p>", "", regex=False)
)

features = []
for _, r in df.iterrows():
    if pd.isna(r["longitude"]) or pd.isna(r["latitude"]):
        continue

    features.append({
        "type": "Feature",
        "geometry": {
            "type": "Point",
            "coordinates": [float(r["longitude"]), float(r["latitude"])]
        },
        "properties": {
            "name_en": r["name_en"],
            "short_description_en": r["short_description_en"],
            "states_name_en": r["states_name_en"],
            "id_no": r["id_no"],,
            "heritage_category": r["category"]
            "danger": int(r["danger"]) if not pd.isna(r["danger"]) else 0
        }
    })

geojson = {
    "type": "FeatureCollection",
    "features": features
}

with open("sites.geojson", "w", encoding="utf-8") as f:
    json.dump(geojson, f, ensure_ascii=False)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

URL = "https://whc.unesco.org/en/list/102/gallery"

opts = Options()
opts.add_argument("--headless=new")
opts.add_argument("--no-sandbox")
opts.add_argument("--disable-dev-shm-usage")
opts.add_argument("--window-size=1400,900")
opts.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36")

def get_picture(URL):
    driver = webdriver.Chrome(options=opts)  # uses chromedriver in PATH
    driver.get(URL)
    
    html = driver.page_source
    driver.quit()
    
    soup = BeautifulSoup(html, "html.parser")
    og = soup.select_one('meta[property="og:image"]')
    print("og:image:", og["content"] if og else "NOT FOUND")
```
